```python
missing_values = data.isnull().sum()
missing_values
```

```python
data_no_cigarettes = data[data['PRICE'] != 0.00]
data_no_cigarettes
```

```python
data.drop(columns='QTY', inplace=True) #all are 1, so it doen't affect anything if we drop the column
```

```python
#show the data just for store 32
data_store_32 = data_no_cigarettes[data_no_cigarettes['STORE'] == 32]
data_store_32

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(3,1, figsize=(10,15))
sns.lineplot(data_store_32, x="WEEK", y="PROFIT", errorbar=None, ax=ax[0])
ax[0].set_title("Profits over Time in Store 32")

sns.lineplot(data_store_32, x="WEEK", y="PRICE",errorbar=None, ax=ax[1])
ax[1].set_title("Profits by Store over time")

sns.lineplot(data_store_32, x="WEEK", y="PROFIT", hue="UPC",errorbar=None, ax=ax[2])
ax[2].set_title("Profits by Product over time")
```
